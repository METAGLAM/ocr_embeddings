```python
import json
import time
import copy
import pandas as pd
from align_paragraphs import Line, write_lines, read_lines, check_line, extract_counts
```

# Experiment 1a
**Accuracy and Error rate of the OCR-correction module in BNC dataset**

**Comparison of methods:**

1. Preprocessing + spelling correction + BERT
2. Preprocessing + spelling correction algorithm by symspellpy: \
    symspellpy link: https://pypi.org/project/symspellpy/ \
    SymSpell link: https://github.com/wolfgarbe/SymSpell

3. Preprocessing + BERT (models extracted from huggingface transformers)

## 1. Symspell + BERT

```python
bloc1 = json.load(open('../data/interim/misspelled_words_jsons/misspelled_words_bloc1_paragraphs_gt.json', 'r'))
bloc2 = json.load(open('../data/interim/misspelled_words_jsons/misspelled_words_bloc2_paragraphs_gt.json', 'r'))
```

#### Align sentences bloc 1

```python
aligned_lines_bloc1 = []
bloc1_raw_misspelled_words = []
bloc1_pro_misspelled_words = []
for i, line in enumerate(bloc1):
    start = time.time()
    print(f"{i+1} de {len(bloc1)} \t {round(i/len(bloc1)*100, 2)}%   \t {len(line['manually_corrected_line'])}", end='\t')

    aligned_line = Line(line)
    aligned_line.join_alignments()
    aligned_lines_bloc1.append(aligned_line)
    
    bloc1_raw_misspelled_words.append(aligned_line.raw_misspelled_words)
    bloc1_pro_misspelled_words.append(aligned_line.pro_misspelled_words)

    print(f"{round(time.time() - start, 2)}s")
```

```python
write_lines(aligned_lines_bloc1, '../data/processed/aligned_lines_jsons/aligned_lines_bloc1.json')
```

```python
loaded_lines_bloc1 = read_lines('../data/processed/aligned_lines_jsons/aligned_lines_bloc1.json')
```

```python
aligned_lines_bloc1_loaded = []
for i, line in enumerate(loaded_lines_bloc1):
    start = time.time()
    print(f"{i+1} de {len(loaded_lines_bloc1)}  ", end='\t')

    aligned_line = Line(load = line)
    aligned_line.join_alignments()
    aligned_lines_bloc1_loaded.append(aligned_line)

    print(f"{round(time.time() - start, 2)}s")
```

```python
for i in range(len(aligned_lines_bloc1_loaded)):
    print(i)
    check_line(aligned_lines_bloc1_loaded, i)
    print()
```

#### Align sentences bloc 2

```python
aligned_lines_bloc2 = []
bloc2_raw_misspelled_words = []
bloc2_pro_misspelled_words = []
for i, line in enumerate(bloc2):
    if line['manually_corrected_line']:
        start = time.time()
        print(f"{i+1} de {len(bloc2)} \t {len(line['manually_corrected_line'])}", end='\t')
    
        aligned_line = Line(line)
        aligned_line.join_alignments()
        aligned_lines_bloc2.append(aligned_line)
    
        bloc2_raw_misspelled_words.append(aligned_line.raw_misspelled_words)
        bloc2_pro_misspelled_words.append(aligned_line.pro_misspelled_words)
    
        print(f"{round(time.time() - start, 2)}s")
```

```python
write_lines(aligned_lines_bloc2, '../data/processed/aligned_lines_jsons/aligned_lines_bloc2.json')
```

```python
loaded_lines_bloc2 = read_lines('../data/processed/aligned_lines_jsons/aligned_lines_bloc2.json')
```

```python
aligned_lines_bloc2_loaded = []
for i, line in enumerate(loaded_lines_bloc2):
    start = time.time()
    print(f"{i+1} de {len(loaded_lines_bloc2)}  ", end='\t')

    aligned_line = Line(load = line)
    aligned_line.join_alignments()
    aligned_lines_bloc2_loaded.append(aligned_line)

    print(f"{round(time.time() - start, 2)}s")
```

```python
for i in range(len(aligned_lines_bloc2_loaded)):
    print(i)
    check_line(aligned_lines_bloc2_loaded, i)
    print()
```

### Extract metrics: accuracy and error rate

```python
raw_correct_bloc1, pro_correct_bloc1, lev_raw_bloc1, lev_pro_bloc1, len_char_gt_bloc1, len_words_gt_bloc1 = \
    extract_counts(aligned_lines_bloc1_loaded)
raw_correct_bloc2, pro_correct_bloc2, lev_raw_bloc2, lev_pro_bloc2, len_char_gt_bloc2, len_words_gt_bloc2 = \
    extract_counts(aligned_lines_bloc2_loaded)
```

```python
print(f'\nMETRICS:\n\n'
      f'Correct words (accuracy) raw bloc1: {round(raw_correct_bloc1 / len_words_gt_bloc1 *100,2)}% \n' \
      f'Correct words (accuracy) pro bloc1: {round(pro_correct_bloc1 / len_words_gt_bloc1 *100,2)}% \n--\n' \
      
      f'Correct words (accuracy) raw bloc2: {round(raw_correct_bloc2 / len_words_gt_bloc2 *100,2)}% \n' \
      f'Correct words (accuracy) pro bloc2: {round(pro_correct_bloc2 / len_words_gt_bloc2 *100,2)}% \n--\n' \

      f'Correct words (accuracy) raw {round((raw_correct_bloc1 + raw_correct_bloc2)/(len_words_gt_bloc1+len_words_gt_bloc2)*100,2)}% \n' \
      f'Correct words (accuracy) pro {round((pro_correct_bloc1 + pro_correct_bloc2)/(len_words_gt_bloc1+len_words_gt_bloc2)*100,2)}% \n--\n' \
      
      f'Error rate raw bloc1: {round(lev_raw_bloc1/len_char_gt_bloc1*100,2)} \n' \
      f'Error rate pro bloc1: {round(lev_pro_bloc1/len_char_gt_bloc1*100,2)} \n--\n' \
      
      f'Error rate raw bloc2: {round(lev_raw_bloc2/len_char_gt_bloc2*100,2)} \n' \
      f'Error rate pro bloc2: {round(lev_pro_bloc2/len_char_gt_bloc2*100,2)} \n--\n' \

      f'Error rate raw: {round((lev_raw_bloc1+lev_raw_bloc2)/(len_char_gt_bloc1+len_char_gt_bloc2)*100,2)} \n' \
      f'Error rate pro: {round((lev_pro_bloc1+lev_pro_bloc2)/(len_char_gt_bloc1+len_char_gt_bloc2)*100,2)} \n')
```
